```python
bpop_math_student_df = 0
arr = []
keys = []
key_value = []

print(len(bpop_math_students))

for student_id in bpop_math_students: # 12
    keys.append(student_id)
    bpop_math_student_df = bpop_math_df[bpop_math_df['Anon Student Id'] == student_id]
    cols = ['Anon Student Id', 'Level (Tutor)', 'Outcome']
    bpop_math_student_df = bpop_math_student_df[cols]
    bpop_math_level_tutor = bpop_math_student_df['Level (Tutor)'].unique()
    print("Types of tutor levels assoc. with this student", student_id, ": ", len(bpop_math_level_tutor))
    arr = []
    
    for tutor_level in bpop_math_level_tutor:
        student_df = bpop_math_student_df[bpop_math_student_df['Level (Tutor)'] == tutor_level]

        num_correct = student_df[student_df['Outcome'] == 'CORRECT'].count()['Anon Student Id']
        num_incorrect = student_df[student_df['Outcome'] == 'INCORRECT'].count()['Anon Student Id']
        num_attempts = num_correct + num_incorrect
        percent_correct = 100*num_correct/num_attempts
        
        subarr = [student_id, tutor_level, num_correct, num_attempts, percent_correct]
        arr.append(subarr)
    
    student_df = pd.DataFrame(arr, columns = ['Anon Student Id', 'Level (Tutor)', '# Correct', '# attempted', '% correct'])
    key_value_pair = (student_id, student_df) 
    key_value.append(key_value_pair)
```

```python
# X and y -> inputs and outputs for the classification model
cols = ['Duration (sec)', 'Level (Tutor)', 'Attempt At Step', 'CF (Matrix Level)', 'CF (Matrix Order)', 'Outcome']
X = bpop_math_df[cols]
X = X.sample(frac=1).reset_index(drop=True)
y = X[['Outcome']]
```

```python
# Clean duration (sec) columns: Convert all values to float, if not able to convert set value in that row as duration_mean
strings = 0
ints = 0
floats = 0
other = 0
cnt = 0
sums = 0
posn = []
X_duration_arr = X['Duration (sec)'].array

for i in range(len(X_duration_arr)):
    if isinstance(X_duration_arr[i], int):
        ints += 1
    elif isinstance(X_duration_arr[i], str):
        strings += 1
        try:
            X_duration_arr[i] = float(X_duration_arr[i])
        except ValueError:
            print("Cant convert to a float", X_duration_arr[i], "Position: ",i)
            posn.append(i)
    elif isinstance(X_duration_arr[i], float):
        floats += 1
    
    if isinstance(X_duration_arr[i], float):
        sums += X_duration_arr[i]
        cnt += 1

duration_mean = sums/cnt
for i in range(len(posn)):
    X_duration_arr[posn[i]] = duration_mean

for i in range(len(X_duration_arr)):
    col = bpop_math_df.columns.get_loc('Duration (sec)')
    val = X_duration_arr[i]
    X.iloc[i, col] = val
```

```python
# Ordinal Encoding of Categorical column 'Level (Tutor)'
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
level_tutor = X['Level (Tutor)'].tolist()
tutor_level_2dlist = []
for i in range(len(level_tutor)):
    tutor_level_2dlist.append([level_tutor[i]])

enc.fit(tutor_level_2dlist)

enc.transform([['bpop.num_0..9.by.1.asc.q2q.AV.rise.12__it_2']])

col = X.columns.get_loc('Level (Tutor)')
num_entries = X.count()['Duration (sec)']
for i in range(num_entries):
    val = X.iloc[i, col]
    val = enc.transform([[val]])[0][0]
    X.iloc[i, col] = val
```

```python
# Split original data X into trainset and testset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Drop 'Outcome' column in X_train and X_test
X_train = X_train.drop(['Outcome'], axis=1)
X_test = X_test.drop(['Outcome'], axis=1)
```

```python
data_final_vars=X_train.columns.values.tolist()
y_=['y']
X=[i for i in data_final_vars if i not in y_]
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
rfe = RFE(logreg, 20)
rfe = rfe.fit(X_train, y_train)
print(rfe.support_)
print(rfe.ranking_)

    
```

```python
# Set incorrect -> 0 and correct to 1 in y_train and y_test

col = y_train.columns.get_loc('Outcome')
for i in range(y_train.count()['Outcome']): 
    val = y_train.iloc[i, col]
    if val == 'INCORRECT':
        val = 0
    else: val = 1
    y_train.iloc[i, col] = val

col = y_test.columns.get_loc('Outcome')
for i in range(y_test.count()['Outcome']): 
    val = y_test.iloc[i, col]
    if val == 'INCORRECT':
        val = 0
    else: val = 1
    y_test.iloc[i, col] = val
```
