### bubble_pop with CF(Matrix) -> math analysis

```python
uniq_bpop_math_level_tutor = bpop_math_df['Level (Tutor)'].unique()
no_of_bpop_math_level_tutor = len(uniq_bpop_math_level_tutor)
print("Types of 'Level (Tutor)': ", no_of_bpop_math_level_tutor)

bpop_math_problem_names = bpop_math_df['Problem Name'].unique()
no_of_bpop_math_prob_names = len(bpop_math_problem_names)
print("Problem names: ", no_of_bpop_math_prob_names)
bpop_math_df
```

```python
bpop_math_student_df = 0
cnt = 0
flag = 1
arr = []
keys = []
key_value = []

print(len(bpop_math_students))

for student_id in bpop_math_students: # 12
    keys.append(student_id)
    bpop_math_student_df = bpop_math_df[bpop_math_df['Anon Student Id'] == student_id]
    cols = ['Anon Student Id', 'Level (Tutor)', 'Outcome']
    bpop_math_student_df = bpop_math_student_df[cols]
    bpop_math_level_tutor = bpop_math_student_df['Level (Tutor)'].unique()
    print("Types of tutor levels assoc. with this student", student_id, ": ", len(bpop_math_level_tutor))
    arr = []
    
    for tutor_level in bpop_math_level_tutor:
        student_df = bpop_math_student_df[bpop_math_student_df['Level (Tutor)'] == tutor_level]

        num_correct = student_df[student_df['Outcome'] == 'CORRECT'].count()['Anon Student Id']
        num_incorrect = student_df[student_df['Outcome'] == 'INCORRECT'].count()['Anon Student Id']
        num_attempts = num_correct + num_incorrect
        percent_correct = 100*num_correct/num_attempts
        
        subarr = [student_id, tutor_level, num_correct, num_attempts, percent_correct]
        arr.append(subarr)
    
    student_df = pd.DataFrame(arr, columns = ['Anon Student Id', 'Level (Tutor)', '# Correct', '# attempted', '% correct'])
    key_value_pair = (student_id, student_df) 
    key_value.append(key_value_pair)
```

```python
map_student_to_perf = dict(key_value)
```

```python
result_df = None

for student_id in bpop_math_students:
#     map_student_to_perf[student_id] is a pandas dataframe
    if result_df is None:
        result_df = map_student_to_perf[student_id]
    else:
        result_df = result_df.append(map_student_to_perf[student_id], ignore_index=True)
    
result_df
```

```python
# create excel writer object
writer = pd.ExcelWriter('output.xlsx')
# write dataframe to excel
result_df.to_excel(writer)
# save the excel
writer.save()
```
