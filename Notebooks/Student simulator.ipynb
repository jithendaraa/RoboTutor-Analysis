```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
pd.options.display.max_rows = 100
```

```python
cta_df = pd.read_excel('Data/CTA.xlsx') 
cta_columns = cta_df.columns.tolist()
cta_columns
kc_to_tutorID_dict = {}
keys = len(cta_columns)

for i in range(keys):
        kc_to_tutorID_dict[cta_columns[i]] = []

col_values = []
tutorIDs = []

for column in cta_columns:
    col_values = cta_df[[column]].values.ravel()
#     col_values = pd.unique(col_values)
    remove_idx = []
    for i in range(len(col_values)):
        
        if(isinstance(col_values[i], str) == False):
            col_values[i] = str(col_values[i])
        
        if col_values[i].lower() == 'nan' or col_values[i].lower()[:6] == 'column':
            remove_idx.append(i)
    
    col_values = np.delete(col_values, remove_idx)
    kc_to_tutorID_dict[column] = col_values
    
    for val in col_values:
        tutorIDs.append(val)

tutorIDs = np.array(tutorIDs)
# 2475
tutorIDs.shape[0]
# for key in kc_to_tutorID_dict:
#     print(key + ":", str((kc_to_tutorID_dict[key])))
```

```python
entries = []

for i in range(len(unique_tutors)):
    tutor_name = unique_tutors[i]
    df = transac_df[transac_df["Level (Tutor)"] == tutor_name]
    correct = df[df["Outcome"] == "CORRECT"].count()[0]
    incorrect = df[df["Outcome"] == "INCORRECT"].count()[0]
    total = correct + incorrect
    student_ids = pd.unique(transac_df[transac_df["Level (Tutor)"] == tutor_name]["Anon Student Id"].values.ravel())
    for i in range(len(student_ids)):
        student_ids[i] = str(student_ids[i])
    student_ids = pd.unique(student_ids).tolist()
    entry = [tutor_name, correct, incorrect, total, str(correct*100.0/total)+"%", len(student_ids)]
    print(entry)
    entries.append(entry)
    
```

```python
difficulty_df = pd.DataFrame(entries, columns=["Level (Tutor)", "num_correct", "num_incorrect", "total_attempts", "%correct", "Students Attempted"])
# difficulty_df.to_excel("difficulty.xlsx", sheet_name='Sheet_name_1')  
difficulty_df
```
