```python
import pandas as pd
import os
import numpy as np
a = np.arange(114, 142).tolist()
```

```python
village_nums = []
for i in range(len(a)):
    village_nums.append(str(a[i]))
```

```python

for village_num in village_nums:
    transac_df = pd.read_csv('hmm-scalable/village_' + str(village_num) + '_KCSubtests.txt', delimiter='\t')

    student_ids = transac_df['Anon Student Id'].values.tolist()
    corrects = transac_df['Outcome'].values.tolist()

    for i in range(len(corrects)):
        correct = corrects[i]
        if correct == 'CORRECT':
            corrects[i] = 1
        elif correct=="INCORRECT":
            corrects[i] = 2
        else:
            print("WEIRD VALUES DETECTED......")
        
    problem_names = transac_df['Problem Name'].values.tolist()
    step_names = transac_df['Step Name'].values.tolist()
    tutor_names = transac_df['Level (Tutor Name)'].values.tolist()
    tutor_name_names = transac_df['Level (Tutor Name)'].values.tolist()
    kc_0 = transac_df['KC(Subtest)'].values.tolist()
    kc_1 = transac_df['KC(Subtest).1'].values.tolist()
    kc_2 = transac_df['KC(Subtest).2'].values.tolist()
    kc_3 = transac_df['KC(Subtest).3'].values.tolist()

    skills = []

    for i in range(len(kc_0)):
        skill = kc_0[i].replace(" ", "")
        if isinstance(kc_1[i], str) == True:
            skill = skill + "~" + kc_1[i].replace(" ", "")
        if isinstance(kc_2[i], str) == True:
            skill = skill + "~" + kc_2[i].replace(" ", "")
        if isinstance(kc_3[i], str) == True:
            skill = skill + "~" + kc_3[i].replace(" ", "")
        skills.append(skill)
    
    step_granularity = []
    problem_granularity = []

    for i in range(len(problem_names)):
        step = "problem:" + problem_names[i].replace(" ", "") + ":step:" + str(step_names[i])
        problem = "problem:" + problem_names[i].replace(" ", "")
        step_granularity.append(step)
        problem_granularity.append(problem)
    
    step_file = open('hmm-scalable/village_'+ str(village_num)+'_step_transac.txt', "w")
    problem_file = open('hmm-scalable/village_' + str(village_num) + '_problem_transac.txt', "w")
    
    step_L = []
    problem_L = []
    for i in range(len(student_ids)):
        step_line = str(corrects[i]) + "\t" + str(student_ids[i]) + "\t" + str(step_granularity[i]) + "\t" + str(skills[i]) + "\n"
        step_L.append(step_line)
        problem_line = str(corrects[i]) + "\t" + str(student_ids[i]) + "\t" + str(problem_granularity[i]) + "\t" + str(skills[i]) + "\n"
        problem_L.append(problem_line)
    
    step_file.writelines(step_L) 
    step_file.close()
    problem_file.writelines(problem_L) 
    problem_file.close()

```
